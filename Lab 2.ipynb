### Insights from inspecting tickets of class 1, 2 & 3

* Class 1 tickets are the most expensive, having an average fare of 84.15, compared with 20.66 for class 2 tickets and 13.77 for class 3 tickets
* Based on the below table, people who bought tickets of class 1 have the most survivors:

* Class 1 tickets has the oldest people, with the 75th percentile age being 48 years. This means they'd likely been working for a while and were therefore richer than the younger passengers who mostly bought Class 2 & Class 3 tickets

### Plots

* The first plot below gives information for each ticket class, showing how many people survived and how many who didn't.

* The second plot gives a better picture of the requirements for this assignment:
    * It shows the fare ranges for those who survived and those who didn't for the 3 fare classes, ignoring all fare prices of zero
    * The people who bought class 1 tickets and survived, which happens to be the class with the highest percentage of survivors (50%), also paid the most as can be seen from the box plot for "Survived" in Pclass = 1
    * Pclass = 3 has the least percentage of survivors (less than 25%) and the survivors there paid much less than the other two classes, with the mean fare of survivors in this ticket class being 9

```python
sns.countplot(x = 'Pclass', hue='Survived', data=df)
```

```python
g = sns.catplot(x="Fare", y="Survived", row="Pclass",
                kind="box", orient="h", height=1.5, aspect=4,
                data=df.query("Fare > 0"))
g.set(xscale="log");
```
